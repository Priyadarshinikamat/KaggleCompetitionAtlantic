```python
import pandas as pd
data= pd.read_csv("/content/train.csv")
data.head()
data.describe(include='all')
```

```python
data['HomePlanet'].value_counts()
```

```python
data['Destination'].value_counts()
```

```python
one_hot_encoded_data = pd.get_dummies(data, columns = ['HomePlanet', 'Destination'])
print(one_hot_encoded_data)
```

```python
for ind in one_hot_encoded_data.index:
  if one_hot_encoded_data['CryoSleep'][ind]==True:
    one_hot_encoded_data['CryoSleep'][ind]= 1
  else:
    one_hot_encoded_data['CryoSleep'][ind]= 0
  if data['Transported'][ind]==True:
    one_hot_encoded_data['Transported'][ind]= 1
  else:
    one_hot_encoded_data['Transported'][ind]= 0
  if data['VIP'][ind]==True:
    one_hot_encoded_data['VIP'][ind]= 1
  else:
    one_hot_encoded_data['VIP'][ind]= 0

print(one_hot_encoded_data)
```

```python
finalsemi= one_hot_encoded_data.drop(['PassengerId','Name','Cabin'],axis=1)
```

```python
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler

#df2 = finalsemi[['CryoSleep', 'VIP','Transported']].copy()
 
scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(finalsemi)
temp2=pd.DataFrame(df_scaled)
#df = pd.concat([temp,df2],axis=1)
print("Scaled Dataset Using MinMaxScaler")
temp = temp2.dropna(axis=0)
temp.head()
```

```python
y = temp[14]
x = temp.drop(14, axis=1)

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, stratify = y, random_state = 7)
```

```python
clf = LogisticRegression(random_state=0).fit(x_train, y_train)
y_pred= clf.predict(x_test)
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score, accuracy_score, roc_auc_score
print("-------------------------TEST SCORES-----------------------") 
print(f"Recall: { round(recall_score(y_test, y_pred)*100, 4) }")
print(f"Precision: { round(precision_score(y_test, y_pred)*100, 4) }")
print(f"F1-Score: { round(f1_score(y_test, y_pred)*100, 4) }")
print(f"Accuracy score: { round(accuracy_score(y_test, y_pred)*100, 4) }")
print(f"AUC Score: { round(roc_auc_score(y_test, y_pred)*100, 4) }")
```
